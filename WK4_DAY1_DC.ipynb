```python
# Hint 1: Use np.random.uniform(low, high, size) to generate the temperature data.
temperatures = np.random.uniform(low=-5, high=35, size=(7, 4))

print("Week temperatures :", temperatures)
```

```python
# Hint 2: Create a DataFrame using pd.DataFrame(data, index, columns) with appropriate index and columns.

# Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.
# Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.
```

```python
# Step 1 : Generate temperatures (10 cities x 12 months)
data = np.random.uniform(-5, 35, size=(10, 12))

# Step 2 : Cities names (index) and months (columns)
cities = [f"City_{i+1}" for i in range(10)]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# Step 3 : Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.
df = pd.DataFrame(data, index=cities, columns=months)

# Display
print(df)
```

2. Data Analysis:
Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).
Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

Calculate the annual average temperature for each city.

Identify the city with the highest and lowest average temperature for the year.

```python
# Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).

# Yearly average calculation :
df["Yearly_average"] = df.mean(axis=1)
print(df)
```

```python
# Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

# city with the highest yearly average temperature
hotest_city = df["Yearly_average"].idxmax()
# city with the lowest  yearly average temperature 
coldest_city = df["Yearly_average"].idxmin()

print("hotest_city :", hotest_city)
print("coldest_city :", coldest_city)
```

```python
# Data visualization

# Deliverables:
    # A Jupyter Notebook containing all the code for data generation, analysis, and visualization.

import matplotlib.pyplot as plt

# Average temperature per month (column-wise mean)
monthly_avg = df[months].mean()

# Plotting
plt.figure(figsize=(12, 6))
monthly_avg.plot(marker='o')

plt.title("Average Monthly Temperature Across All Cities")
plt.xlabel("Month")
plt.ylabel("Temperature (Â°C)")
plt.grid(True)
```

```python
# Transposition as to get months on x axis and cities as lines

# Suppose df is your DataFrame with cities as index and months as columns
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
          
df[months].T.plot(figsize=(12, 6), marker='o')

# Plot fine tuning 
plt.xticks(ticks=range(12), labels=months)  # Ensure that all months are displayed 
plt.legend(title="Cities", bbox_to_anchor=(1.05, 1), loc='upper left')  # Legend display

plt.tight_layout()
plt.show()
```
