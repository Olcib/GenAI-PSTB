Flip an image horizontally and then vertically using mirror and flip functions from ImageOps
Zoom in on an image (scale by 1.2x) using .resize(...)

```python
from PIL import Image, ImageOps

# Image opening:
img = Image.open("00_001.png")
```

```python
# Horizontal mirror (flip left to right)
horizontally_flipped = ImageOps.mirror(img)
```

```python
# Vertical Mirror (flip up-bottom)
vertically_flipped = ImageOps.flip(img)
```

```python
# Zoom (1.2x)
width, height = img.size
zoomed = img.resize((int(width * 1.2), int(height * 1.2)))
```

```python
# Results display :
horizontally_flipped.show(title="Horizontal Flip")
vertically_flipped.show(title="Vertical Flip")
zoomed.show(title="Zoom x1.2")
```
